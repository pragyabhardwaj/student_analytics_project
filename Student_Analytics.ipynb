```
from pyspark import SparkConf, SparkContext
conf = SparkConf().setAppName("Average")
sc = SparkContext.getOrCreate(conf = conf)
```

```
rdd = sc.textFile("/FileStore/tables/Student_Analytics.csv")
headers = rdd.first()
rdd = rdd.filter(lambda x: x!=headers)
rdd = rdd.map(lambda x:x.split(','))
#rdd.collect()
```

```
rdd.count()  ##number of students in the file
```

```
##total marks achieved by female and male students
rdd2 = rdd.map(lambda x:(x[1],int(x[5])))
rdd2=rdd2.reduceByKey(lambda x,y:x+y)
rdd2.collect()
```

```
##total number of students that have passed or failed, 50+ marks are required to pass the course
pass_students_count=rdd.filter(lambda x:int(x[5])>50).count()
print("The pass students count is: ", pass_students_count)
failed_students_count = rdd.filter(lambda x:int(x[5])<=50).count()
print("The failed students count is: ", failed_students_count)
```

```
##show the total number of students enrolled per course
rdd3=rdd.map(lambda x:(x[3],1))
rdd3=rdd3.reduceByKey(lambda x,y:x+y)
rdd3.collect()
```

```
##total marks achieved per course
rdd4 = rdd.map(lambda x:(x[3],int(x[5])))
#rdd4
rdd4=rdd4.reduceByKey(lambda x,y:x+y)
rdd4.collect()
```

```
## Average marks achieved per course
rdd5 = rdd
rdd5 = rdd5.map(lambda x: (x[3],(int(x[5]),1)))
rdd5.collect()
```

```
##fetching the sum of marks per student and total count of students per course
rdd5=rdd5.reduceByKey(lambda x,y:(x[0]+y[0], x[1]+y[1]))
#rdd5.map(lambda x: (x[0],(x[1][0]/x[1][1]))).collect()
rdd5.collect()
```

```
## now find the average
rdd5.map(lambda x:(x[0],(x[1][0]/x[1][1]))).collect()
```

```
rdd5.mapValues(lambda x: (x[0],x[1])).collect()
```

```
rdd5.mapValues(lambda x: (x[0]/x[1])).collect()
```

```
##minimum and maximum values achieved per course
rdd6 = rdd
rdd6 = rdd6.map(lambda x: (x[3],int(x[5])))
#rdd6.collect()
print(rdd6.reduceByKey(lambda x,y: x if x>y else y).collect())
print(rdd6.reduceByKey(lambda x,y: x if x<y else y).collect())
```
